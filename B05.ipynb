```python
from PIL import Image
from imgtools import *
from IPython.display import display
```

```python
my_path = 'D:/tgmt_img/img03.jpg'
img = load_image(my_path)
display(img)
```

```python
rotated_image = img.rotate(90)
display(rotated_image)
```

```python
rotated_image = rotated_image.rotate(180)
display(rotated_image)
```

```python
transposed_image = img.transpose(Image.Transpose.ROTATE_90)
display(transposed_image)
```

```python
transposed_image = transposed_image.transpose(Image.Transpose.ROTATE_180)
display(transposed_image)
```
