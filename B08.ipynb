```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 40, 30]

x1 = [5, 4, 3, 2, 1]
y1 = [35, 25, 10, 5, 40]

plt.plot(x, y, 'bo-', label='Dữ liệu')
plt.plot(x1, y1, marker='o', color='r', linestyle='-', label='Dữ liệu-1')
plt.legend()
plt.xlabel('Trục x')
plt.ylabel('Trục y')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 10, 15, 24, 13]

plt.bar(categories, values, color='g', alpha=0.6)
plt.xlabel('Danh mục')
plt.ylabel('Giá trị')
plt.show()
```

```python
my_labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
my_colors = ['red', 'green', 'blue', 'yellow']

plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.1f%%')
plt.title('Biểu đồ hình tròn')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

plt.scatter(x, y, c='r', marker='o', label='Dữ liệu')
plt.title('Biểu đồ phân tán ví dụ')
plt.xlabel('Trục x')
plt.ylabel('Trục y')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [15, 18, 22, 30, 35, 45, 50, 55, 65]

plt.boxplot(data)
plt.title('Biểu đồ hộp ví dụ')
plt.ylabel('Giá trị')
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [15, 18, 22, 30, 35, 45, 50, 55, 65]

plt.violinplot(data)
plt.title('Biểu đồ violin ví dụ')
plt.ylabel('Giá trị')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 4, 3, 7, 6]

N = len(categories)
values += values[:1]

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, values, 'o-', linewidth=2)
plt.fill(angles, values, 'b', alpha=0.1)
plt.title('Biểu đồ radar ví dụ')
plt.show()
```
